# Line Plots

```python
import matplotlib.pyplot as plot

import numpy as np

%matplotlib inline
```

```python
x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
```

```python
plot.plot(x,y)
plot.show()
```

# Saving Plot

```python
x = [0,2,4,6]
y = [1,3,4,6]

plot.plot(x,y)
plot.xlabel('x values')
plot.ylabel('y values')
plot.title('plotted x and y values')
plot.legend(['Data 1'])
```

```python
x = [0,2,4,6]
y = [1,3,4,6]

plot.plot(x,y)
plot.xlabel('x values')
plot.ylabel('y values')
plot.title('plotted x and y values')
plot.legend(['Data 1'])

#Saving the plots
plot.savefig('plot1.png', dpi = 350, bbox_inches = 'tight')

plot.show()
```

# Multi Line Plots

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plot.subplots()

ax.plot(x,y)
ax.plot(x,z)


plot.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plot.subplots()

ax.plot(x,y)
ax.plot(x,z)

ax.set_title("Hello")
ax.legend(['sin', 'cos'])
ax.xaxis.set_label_text('Angle')
ax.yaxis.set_label_text('Sine and cosine')


plot.show()
```

# Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')

mu = 80
sigma = 7
x = np.random.normal(mu, sigma, size = 200)

fig, ax = plt.subplots()

ax.hist(x, 20)
ax.set_title("New Histogram")
ax.set_xlabel("Bin range")
ax.set_ylabel("frequency")

fig.tight_layout()
plt.show()
```
